```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fonction image_load (version améliorée)
def image_load(image_path):
    rgb_image = cv2.imread(image_path)
    if rgb_image is None:
        raise FileNotFoundError(f"Erreur : Impossible de charger l'image {image_path}")
    rgb_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB)
    gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if gray_image is None:
        raise FileNotFoundError(f"Erreur : Impossible de charger l'image {image_path} en niveaux de gris")

    if len(rgb_image.shape) == 3 and rgb_image.shape[2] == 3:
        if np.array_equal(rgb_image[...,0], rgb_image[...,1]) and np.array_equal(rgb_image[...,1], rgb_image[...,2]):
            return gray_image, 'Gray'
        else:
            return rgb_image, 'RGB'
    else:
        return gray_image, 'Gray'
```

```python
# Fonctions apply_convolution et convolve_channel (comme précédemment définies)
def convolve_channel(image, kernel):
    assert image.shape[0] >= kernel.shape[0], "L’image est trop petite pour le noyau en hauteur"
    assert image.shape[1] >= kernel.shape[1], "L’image est trop petite pour le noyau en largeur"

    img_height, img_width = image.shape
    kernel_height, kernel_width = kernel.shape
    pad_height, pad_width = kernel_height // 2, kernel_width // 2

    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant')

    output = np.zeros_like(image, dtype=np.float32)

    for i in range(img_height):
        for j in range(img_width):
            region = padded_image[i:i + kernel_height, j:j + kernel_width]
            output[i, j] = np.sum(region * kernel)

    assert output.shape == image.shape, "La taille de la sortie doit être égale à celle de l'image d'entrée"

    return output
```

```python
def apply_convolution(image, kernel, image_type):
    assert isinstance(image, np.ndarray), "L’image doit être un tableau NumPy"
    assert isinstance(kernel, np.ndarray), "Le noyau doit être un tableau NumPy"
    assert len(kernel.shape) == 2, "Le noyau doit être une matrice 2D"
    assert kernel.shape[0] == kernel.shape[1], "Le noyau doit être carré"
    assert kernel.shape[0] % 2 == 1, "Le noyau doit avoir une taille impaire"
    assert image_type in ['Gray', 'RGB'], "Le type d'image doit être 'Gray' ou 'RGB'"

    if image_type == 'RGB':
        assert len(image.shape) == 3 and image.shape[2] == 3, "L’image RGB doit avoir 3 canaux"
        height, width, channels = image.shape
        output = np.zeros_like(image, dtype=np.float32)
        for c in range(channels):
            output[:, :, c] = convolve_channel(image[:, :, c], kernel)
    else:
        assert len(image.shape) == 2, "L’image en niveaux de gris doit être 2D"
        output = convolve_channel(image, kernel)

    output = np.clip(output, 0, 255).astype(np.uint8)

    assert output.shape == image.shape, "La taille de la sortie doit être égale à celle de l'image d'entrée"

    return output
```

```python
# Fonction d'affichage
def display_images(original, filtered_images, titles):
    n = len(filtered_images) + 1
    plt.figure(figsize=(5 * n, 5))
    plt.subplot(1, n, 1)
    plt.title("Image originale")
    plt.imshow(original, cmap='gray' if len(original.shape) == 2 else None)
    plt.axis('off')

    for i, (img, title) in enumerate(zip(filtered_images, titles), start=2):
        plt.subplot(1, n, i)
        plt.title(title)
        plt.imshow(img, cmap='gray' if len(img.shape) == 2 else None)
        plt.axis('off')

    plt.show()
```

```python

# Définition des filtres
blur_kernel = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9]
])

sobel_horizontal = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

sobel_vertical = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
])

sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

edge_detection_kernel = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
])

np.random.seed(1)
random_kernel_3 = np.random.randn(3, 3)
random_kernel_3 -= random_kernel_3.mean()

np.random.seed(2)
random_kernel_5 = np.random.randn(5, 5)
random_kernel_5 -= random_kernel_5.mean()

np.random.seed(3)
random_kernel_7 = np.random.randn(7, 7)
random_kernel_7 -= random_kernel_7.mean()
```

```python
# Chargement de l'image
image_path = 'cameraman.pgm'  # Remplace par le chemin de ton image
image, image_type = image_load(image_path)

# Préparer les images gris et RGB selon le type détecté
if image_type == 'Gray':
    gray_image = image
    rgb_image = None
else:
    rgb_image = image
    gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY)

# Application des filtres sur gris
blurred_gray = apply_convolution(gray_image, blur_kernel, 'Gray')
sobel_h_gray = apply_convolution(gray_image, sobel_horizontal, 'Gray')
sobel_v_gray = apply_convolution(gray_image, sobel_vertical, 'Gray')
sharpen_gray = apply_convolution(gray_image, sharpen_kernel, 'Gray')
edge_gray = apply_convolution(gray_image, edge_detection_kernel, 'Gray')
random_3_gray = apply_convolution(gray_image, random_kernel_3, 'Gray')
random_5_gray = apply_convolution(gray_image, random_kernel_5, 'Gray')
random_7_gray = apply_convolution(gray_image, random_kernel_7, 'Gray')

# Application des filtres sur RGB si disponible
if rgb_image is not None:
    blurred_rgb = apply_convolution(rgb_image, blur_kernel, 'RGB')
    sobel_h_rgb = apply_convolution(rgb_image, sobel_horizontal, 'RGB')
    sobel_v_rgb = apply_convolution(rgb_image, sobel_vertical, 'RGB')
    sharpen_rgb = apply_convolution(rgb_image, sharpen_kernel, 'RGB')
    edge_rgb = apply_convolution(rgb_image, edge_detection_kernel, 'RGB')
    random_3_rgb = apply_convolution(rgb_image, random_kernel_3, 'RGB')
    random_5_rgb = apply_convolution(rgb_image, random_kernel_5, 'RGB')
    random_7_rgb = apply_convolution(rgb_image, random_kernel_7, 'RGB')
```

```python
# Affichage résultats gris
display_images(gray_image,
               [blurred_gray, sobel_h_gray, sobel_v_gray, sharpen_gray, edge_gray, random_3_gray, random_5_gray, random_7_gray],
               ["Flou", "Sobel Horizontal", "Sobel Vertical", "Netteté", "Contours", "Aléatoire 3x3", "Aléatoire 5x5", "Aléatoire 7x7"])

# Affichage résultats RGB si disponible
if rgb_image is not None:
    display_images(rgb_image,
                   [blurred_rgb, sobel_h_rgb, sobel_v_rgb, sharpen_rgb, edge_rgb, random_3_rgb, random_5_rgb, random_7_rgb],
                   ["Flou", "Sobel Horizontal", "Sobel Vertical", "Netteté", "Contours", "Aléatoire 3x3", "Aléatoire 5x5", "Aléatoire 7x7"])
```
